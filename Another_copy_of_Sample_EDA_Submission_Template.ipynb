## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#check missing values
df.isnull().sum()

#remove duplicate values
df.drop_duplicates()

#check duplicates again
df.duplicated().sum()

````

````python
#coverting rating to numeric
df['rating']=pd.to_numeric(df['rating'],errors='coerce')

#coverting rating_count to numeric
df['rating_count']=df['rating_count'].astype(str)
df['rating_count']=df['rating_count'].str.replace(',','')
df['rating_count']=pd.to_numeric(df['rating_count'],errors='coerce')
````

**Top 10 Product Categories:**

````python
# Count top 10 product categories
top_category = df["category"].value_counts().head(10)
plt.figure(figsize=(10,6))

# Create bar chart
plt.barh(top_category.index, top_category.values)
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
# Rotate category names
plt.xticks(rotation=90)
plt.show()
````
